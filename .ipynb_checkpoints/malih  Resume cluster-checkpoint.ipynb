```python
def remove_numbers(documents):
    return [''.join([i for i in doc if not i.isdigit()]) for doc in documents]

def remove_strange_char(documents):
    return [re.sub(r'[^a-zA-Z0-9\#_+-]',r' ', doc) for doc in documents]

def make_tokenize(documents):
    return [doc.lower().split() for doc in documents]

def remove_stopwords(documents):
    stopset = stopwords.words('english')
    return [[word for word in doc if word not in stopset] for doc in documents]

def lemmetization(documents):
    lemm=Mystem()
    return [[''.join(lemm.lemmatize(word)).replace('\n', '') for word in doc] for doc in documents]

def remove_nan(documents):
    return [[word for word in doc if word != "nan"] for doc in documents]
```

```python
docs = remove_strange_char(docs)
docs
```

```python
docs = remove_numbers(docs)
docs
```

```python
docs = make_tokenize(docs)
docs
```

```python
corpus = [' '.join(doc) for doc in docs]

def dummy_fun(doc):
    return doc

vectorizer = TfidfVectorizer(ngram_range=(1,1),
                            max_df=0.99,
                            min_df=0.05,
                            tokenizer=dummy_fun,
                            token_pattern=None,
                            preprocessor=dummy_fun)

X = vectorizer.fit_transform(docs)
```

```python
X
```

```python
def get_umap(data, c=None, n_neighbors=4, min_dist=0.2, n_components=2, metric='cosine', title=''):
    fit = umap.UMAP(
    n_neighbors=n_neighbors,
    min_dist=min_dist,
    n_components=n_components,
    metric=metric,
    )
    u = fit.fit_transform(data)
    return u

data = X.toarray()
umap = get_umap(data)

new_df = pd.DataFrame()

new_df['umap_0'] = umap[:,0]
new_df['umap_1'] = umap[:,1]
```

```python
new_df
```

```python
plt.figure(figsize=(10,7))
sns.lmplot(x="umap_0",
          y="umap_1",
          data=new_df,
          fit_reg=False,
          legend=True,
          height=10,
          aspect=1.5,
          palette="gist_ncar",
          scatter_kws={'edgecolor':'k',
                      's':80})
```

```python
Z = shc.linkage(data, method='ward')

plt.figure(figsize=(15, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
dend = shc.dendrogram(Z)
```

```python
from scipy.cluster.hierarchy import fcluster

k=4
temp = fcluster(Z, k, criterion='maxclust')

new_df['hierarchy'] = temp

df_tf = pd.DataFrame(data, columns=vectorizer.get_feature_names())
df_tf['hierarchy'] = temp

plt.figure(figsize=(10,7))
sns.lmplot(x="umap_0",
          y="umap_1",
          data=new_df,
          fit_reg=False,
          hue='hierarchy',
          legend=True,
          height=10,
          aspect=1.5,
          palette="gist_ncar",
          scatter_kws={'edgecolor':'k',
                      's':80})
```

```python
def perc_func(x, cv_id, h):
    if x.name != cv_id:
        return (x - h[cv_id]).fillna(0)
    else:
        return x
    
def get_matching(cv_id, n_similar, single_tree, relative=False):
    neighbors = single_tree.get_nns_by_vector(data[cv_id], n=n_similar)
    
    table = df_tf.loc[neighbors].drop(['hierarchy'],1).T.sort_values(cv_id, ascending=False)
    indexes = set()
    
    for cv in neighbors:
        indexes.update(set(table.sort_values(cv, ascending=False).index[:50].tolist()))
        
    h = table.loc[indexes].sort_values(cv_id, ascending=False)
    
    info = h.apply(lambda x: perc_func(x, cv_id, h)) if relative else h
    
    info = (info * 100).round(1).style.bar(align='mid', color=['#d65f5f', '#5fba7d'])
    
    return neighbors, info

def show_matching(cv_id, n_similar=1, metric='euclidean'):
    num_features = data.shape[1]
    single_tree = AnnoyIndex(num_features, metric=metric)
    
    i=0
    for vector in data:
        single_tree.add_item(i, vector)
        
        i += 1
    
    single_tree.build(1)
    
    neighbors, info = get_matching(cv_id, n_similar, single_tree, relative=False)
    return neighbors, info
```

```python
res_id = 3
n_similar = 2

show_matching(res_id, n_similar=n_similar, metric='euclidean')[1]
```
